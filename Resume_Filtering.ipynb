Extracting Yes and No from the availability column, because No has many reasons and therefore not possible to give every reason a designated number.

```python
def extract_Yes_No(text):
    Yes_No = re.findall(r'\b(?:Yes|No)\b', text, flags=re.IGNORECASE)
    return ' '.join(Yes_No)

df_cleaned['Availabity'] = df_cleaned['Availability'].apply(extract_Yes_No)
```

Differentiating the main skills needed for the Internship and getting them into new column which counts total needed or desired skills and then normalizing those values.

```python
specific_skills = ['data science', 'data engineering', 'big data', 'power bi', 'tableau','data analytics','big data','data structure','r programming','internet of things (iot)','image processing','statistics','statistical modeling','robotic process automation (rpa)','computervision','opencv']

df_cleaned['Count'] = 0

for index, skill_list in enumerate(df_cleaned['Other skills']):
    count = 0
    if not pd.isnull(skill_list):
        for skill in specific_skills:
            count += skill_list.lower().count(skill)
    df_cleaned.at[index, 'Count'] = count
```

Now, making a new column named as Impact_factor and giving different weights to different factors and calculating impact that a person is making through his/her resume, and giving highest importance to availabilty precedded by skills_count then four main skills and then least importance to Performance in UG and PG.

```python
def calculate_and_store_sum(df):
    df_cleaned['Impact_Factor'] = 0.0
    for index, row in df_cleaned.iterrows():
        row_sum = 0.0
        for col_name, multiplier in zip(df_cleaned.columns, [0.075,0.075,0.2,0.125,0.125,0.125,0.125,0.15]):
            value = row[col_name] * multiplier
            row_sum += value
        df_cleaned.at[index, 'Impact_Factor'] = row_sum

    return df_cleaned
```

```python
df_cleaned = calculate_and_store_sum(df_cleaned)
```

Finding Top Resume's through this function.

```python
def find_top_n_max_values(df, column_name, N):
    top_n_max = df.nlargest(N, column_name)
    top_n_values = []
    for index, row in top_n_max.iterrows():
        value = row[column_name]
        top_n_values.append((index, value))

    return top_n_values
```

These are the Top 10 Resume's from Dataset.

```python
top_n_max_values = find_top_n_max_values(df_cleaned, 'Impact_Factor',10)
for index, value in top_n_max_values:
    print(f"Index: {index}, Value: {value}")
```
